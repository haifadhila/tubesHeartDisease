### Split Data dengan K-Fold

```python
def train_k_fold(k, model_type, X, y):
    # define metrics here, can be accuracy, error, etc.
    # the metrics will be accumulated and the average/variance will be calculated for k times
    kf = KFold(k)
    accuracies = []
    precisions = []
    recalls = []
    for train_idx, test_idx in kf.split(X):
        print(train_idx)
        model = None
        if (model_type=="nb"):
            model = GaussianNB()
        elif(model_type=="dtc"):
            model = DecisionTreeClassifier()
        elif(model_type=="knn"):
            model = KNeighborsClassifier()
        elif(model_type=="mlp"):
            model = MLPClassifier(max_iter=1000)

        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        #Train model here and accumulate metrics 
        model.fit(X_train,y_train)
        iris_predict = model.predict(X_test)
        accuracy = accuracy_score(y_pred=iris_predict, y_true=y_test)
        precision = precision_score(y_pred=iris_predict, y_true=y_test, average='macro')
        recall = recall_score(y_pred=iris_predict, y_true=y_test, average='mcro')

        accuracies.append(accuracy)
        precisions.append(precision)
        recalls.append(recall)
        
    avg_accuracy = sum(accuracies)/len(accuracies)
    avg_precision = sum(precisions)/len(precisions)
    avg_recall = sum(recalls)/len(recalls)
    
    return avg_accuracy, avg_precision, avg_recall

def k_fold_train(k, model_type, X, y) :
    kf = KFold(k, shuffle = True)
    model = None
    # Container
    accuracies = []
    precisions = []
    recalls = []
    
    if model_type == 'nb' :
        model = GaussianNB()
    elif model_type == 'tree' :
        model = DecisionTreeClassifier()
    elif model_type == 'knn' :
        model = KNeighborsClassifier()
    else :
        model = MLPClassifier(max_iter=1000)
    
    for train_idx, test_idx in kf.split(X) :
        x_train, x_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        model.fit(x_train, y_train)
        
        accuracies.append(accuracy_score(y_test, model.predict(x_test)))
        precisions.append(precision_score(y_test, model.predict(x_test), average = 'macro'))
        recalls.append(recall_score(y_test, model.predict(x_test), average = 'macro'))
        
        avg_accuracies = np.mean(accuracies)
        avg_precisions = np.mean(precisions)
        avg_recalls = np.mean(recalls)
        return avg_accuracies, avg_precisions, avg_recalls
```

#### Naive Bayes Model

```python
avg_accuracy,avg_precision, avg_recall = k_fold_train(10, 'nb', train_data_X.values, train_data_y.values)

print("Naive Bayes average accuracy : ", avg_accuracy)
print("Naive Bayes average precision : ", avg_precision)
print("Naive Bayes average recall : ", avg_recall)
```
